```python
Q5: What is Bernaulli Distribution? Give an Example. What is the difference between Bernoulli
Distribution and Binomial Distribution?
Ans-

The Bernoulli distribution is a discrete probability distribution that represents the outcomes of a single 
experiment that can have one of two possible outcomes, often referred to as "success" and "failure."

An example of the Bernoulli distribution is the outcome of a coin toss. If we define "heads" as success and 
"tails" as failure, then the probability of success is p = 0.5, and the probability of failure is 1 - p = 0.5.

 the main differences between Bernoulli distribution and Binomial distribution:
1.Number of Trials:
->Bernoulli distribution models the outcome of a single experiment with two possible outcomes (success or failure).
->Binomial distribution models the number of successes in a fixed number of independent Bernoulli trials.
2.Parameters:
->Bernoulli distribution is characterized by a single parameter p, which represents the probability of success in a single trial.
->Binomial distribution is characterized by two parameters: n, the number of trials, and p, the probability of success in each trial.
3.Probability Mass Function (PMF):
->Bernoulli distribution has a PMF of P(X=x) = p^x * (1-p)^(1-x), where x can take on the values 0 or 1, depending
 on whether the outcome is a failure or success, respectively.
->Binomial distribution has a PMF of P(X=k) = (n choose k) * p^k * (1-p)^(n-k), where k is the number of successes,
 (n choose k) is the binomial coefficient, which represents the number of ways to choose k items from a set of n items, 
and p is the probability of success in each trial.
4.Relationship:
->Bernoulli distribution is a special case of the binomial distribution where n=1.
->Binomial distribution can be derived by summing up n independent and identically distributed Bernoulli trials.
```
